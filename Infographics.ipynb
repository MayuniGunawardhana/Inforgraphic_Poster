```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Read the CSV file into a Pandas DataFrame
df = pd.read_csv('departures_for_foreign_employment_through_private_sources_&_through_agencies_1997_2012_0 (2).csv')

# Print the DataFrame
df.head(5)
```

```python
df.info()
```

```python
sns.lineplot(data=df, x="Year", y="SelfBasisMale ", label="Self Basis Male")
sns.lineplot(data=df, x="Year", y="SelfBasisFemale", label="Self Basis Female")
plt.ylabel("Self Basis Count")
plt.legend()
# Remove the grid
plt.grid(False)

plt.show()
```

```python
sns.lineplot(data=df, x="Year", y="ThroughAgenciesMale", label="Through Agencies Male")
sns.lineplot(data=df, x="Year", y="ThroughAgenciesFemale ", label="Through Agencies Female")
plt.ylabel("Count Through Agencies")
plt.legend()
# Remove the grid
plt.grid(False)

plt.show()
```

```python
print(df.columns)
```

```python
import matplotlib.pyplot as plt
sns.lineplot(data=df, x="Year", y="TotalMale ", label="Total Male")
sns.lineplot(data=df, x="Year", y="TotalFemale", label="Total Female")
plt.ylabel("Count")
plt.legend()
# Remove the grid
plt.grid(False)

plt.show()
```

```python
# Calculate male to female percentages
df['Male_Percentage'] = df['TotalMale '] / (df['TotalMale '] + df['TotalFemale']) * 100
df['Female_Percentage'] = df['TotalFemale'] / (df['TotalMale '] + df['TotalFemale']) * 100

# Plotting
plt.plot(df['Year'], df['Male_Percentage'], label='Male %')
plt.plot(df['Year'], df['Female_Percentage'], label='Female %')

plt.xlabel('Year')
plt.ylabel('Male to Female Percentage')
plt.title('Male to Female Percentage Over the Years')
# Remove the grid
plt.grid(False)

plt.legend()

plt.show()
```

```python
# Calculate male to female percentages
df['Male_Percentage'] = df['TotalMale '] / (df['TotalMale '] + df['TotalFemale']) * 100
df['Female_Percentage'] = df['TotalFemale'] / (df['TotalMale '] + df['TotalFemale']) * 100

# Display the DataFrame
print(df[['Year','Male_Percentage','Female_Percentage']])
```

```python
df_total = df[["TotalMale ", "TotalFemale"]]

# Plotting
#plt.figure(figsize=(10, 7), facecolor='none')  # Setting facecolor to 'none' for transparent background
plt.pie(df_total.sum(), labels=df_total.columns, autopct="%1.1f%%")
plt.title("Total Male vs Total Female")

# Save as PNG
#plt.savefig('total_male_vs_female.png', transparent=True)
```

```python
df_self_basis = df[["SelfBasisMale ", "SelfBasisFemale"]]

plt.figure(figsize=(10, 7), facecolor='none')  # Setting facecolor to 'none' for transparent background
plt.pie(df_self_basis.sum(), labels=df_self_basis.columns, autopct="%1.1f%%")
plt.title("Self Basis Male vs Self Basis Female")

# Save as PNG
plt.savefig('Self Basis Male vs Self Basis Female.png', transparent=True)
plt.show()
```

```python
import matplotlib.pyplot as plt
df_through_agencies = df[["ThroughAgenciesMale", "ThroughAgenciesFemale "]]

plt.figure(figsize=(10, 7), facecolor='none')  # Setting facecolor to 'none' for transparent background
plt.pie(df_through_agencies.sum(), labels=df_through_agencies.columns, autopct="%1.1f%%")
plt.title("Through Agencies Male vs Through Agencies Female")

# Save as PNG
plt.savefig('Through Agencies Male vs Through Agencies Female.png', transparent=True)
plt.show()
```

```python
import matplotlib.pyplot as plt
df['ThroughAgenciesTotal'] = df['ThroughAgenciesMale'] + df['ThroughAgenciesFemale ']
df['SelfBasisTotal'] = df['SelfBasisMale '] + df['SelfBasisFemale']

df_pie = df[['ThroughAgenciesTotal', 'SelfBasisTotal']]

plt.figure(figsize=(10, 7))
plt.pie(df_pie.sum(), labels=df_pie.columns, autopct="%1.1f%%", colors=['lightpink', 'lightblue'])
plt.title("Through Agencies Total vs Self Basis Total")
plt.show()

# Save as PNG
#plt.savefig('Through Agencies Total vs Self Basis Total.png', transparent=True)
```

```python
sns.lineplot(data=df, x="Year", y="ThroughAgenciesTotal", color='lightpink', label="Through Agencies Total")
sns.lineplot(data=df, x="Year", y="SelfBasisTotal ", color='lightblue', label="Self Basis Total")
plt.legend()
# Remove the grid
plt.grid(False)
plt.show()
```
